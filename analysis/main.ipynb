```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
```

```python
df= pd.read_json('./data/22_01_2023/merged_18_22_01_2023.json', orient='records')
df
```

```python
houses = df[df['proper_type'] == 'house']
houses
```

```python
ax = houses['province'].value_counts().plot(kind='bar',
                                    figsize=(14,2),
                                    title="Ads per province")
ax.set_xlabel("Province")
ax.set_ylabel("Ads count")
plt.show()
```
